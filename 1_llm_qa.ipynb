```python
from langgraph.graph import StateGraph,START,END
from langchain_groq import ChatGroq

from typing import TypedDict
from dotenv import load_dotenv

load_dotenv()
```

```python
model = ChatGroq(
            model="llama-3.1-8b-instant",
            temperature=0.0,
            max_retries=2,
        
        )
```

```python
# create a state

class LLMQAState(TypedDict):
    question: str
    answer: str
```

```python
def llm_qa(state: LLMQAState) -> LLMQAState:
    
    #extract question from state
    question = state['question']
    
    #form a prompt for the model
    prompt = f"Answer the following question: {question}"
    # ask that question to the LLM 
    answer = model.invoke(prompt).content
    # update the answer in the state
    
    state['answer'] = answer
    
    return state
    
```

```python
# create a graph

graph = StateGraph(LLMQAState)

# add a node to the graph
graph.add_node('llm_qa', llm_qa)

#add edge to the graph
graph.add_edge(START, 'llm_qa')
graph.add_edge('llm_qa', END)

workflow = graph.compile()
```

```python
# execute the graph with user input
user_question = input("Enter your question for the LLM: ")
initial_state = {'question': user_question}

final_state = workflow.invoke(initial_state)
print(final_state['answer'])
```
